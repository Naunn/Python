```python
df = pd.read_csv('insurance.csv')
```

```python
df['sex'] = df['sex'].apply(lambda x: 1 if x == 'male' else 0)
df['smoker'] = df['smoker'].apply(lambda x: 1 if x == 'yes' else 0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.count()
```

```python
df['southwest'] = df['region']
df['southeast'] = df['region']
df['northwest'] = df['region']
df['northeast'] = df['region']
```

```python
df.head(10)
```

```python
df['southwest'] = df['southwest'].apply(lambda x: 1 if x == 'southwest' else 0)
df['southeast'] = df['southeast'].apply(lambda x: 1 if x == 'southeast' else 0)
df['northwest'] = df['northwest'].apply(lambda x: 1 if x == 'northwest' else 0)
df['northeast'] = df['northeast'].apply(lambda x: 1 if x == 'northeast' else 0)
```

```python
df.head(10)
```

```python
df = df.drop(['region'], axis = 1)
```

```python
df.columns
```

```python
df = df[['age', 'sex', 'bmi', 'children', 'smoker', 'southwest',
       'southeast', 'northwest', 'northeast', 'charges']]
```

```python
df.describe()
```

```python
plt.subplots(figsize=(20,15))
sns.heatmap(df.corr(),annot=True)
```

```python
X = df[['age', 'bmi', 'smoker']]
```

```python
y = df['charges']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm = LinearRegression()
```

```python
lm.fit(X_train,y_train)
```

```python
print(lm.intercept_) # expected mean value of Y when all X=0
```

```python
cdf = pd.DataFrame(lm.coef_,X.columns,columns=['Coeff'])
cdf
```

```python
predictions = lm.predict(X_test)
predictions
```

```python
plt.subplots(figsize=(20,15))
plt.scatter(y_test,predictions)
plt.plot(y_test, y_test, color='grey', linewidth=3)
```

```python
sns.distplot((y_test-predictions))
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test,predictions) # MAE
```

```python
np.sqrt(metrics.mean_squared_error(y_test,predictions)) # RMSE
```

```python
X = df[['age', 'bmi', 'charges']]
```

```python
y = df['smoker']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix
```

```python
rfc = RandomForestClassifier(n_estimators=200)
```

```python
rfc.fit(X_train,y_train)
```

```python
rfc_pred = rfc.predict(X_test)
```

```python
print(confusion_matrix(y_test,rfc_pred))
print('\n')
print(classification_report(y_test,rfc_pred))
```

```python
df['smoker'].value_counts()
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
predictionst = dtree.predict(X_test)
```

```python
print(confusion_matrix(y_test,predictionst))
print('\n')
print(classification_report(y_test,predictionst))
```
