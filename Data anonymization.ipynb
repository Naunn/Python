```python
cdf = pd.read_csv('clean_Salaries.csv')
cdf.head()
```

```python
#We can see that some names are repeating themself. Reason of that may be using data 2011-2014 year.
#What we will do is leave names and only first letter of surname.
```

```python
cdf['EmployeeName'].unique()
```

```python
cdf['EmployeeName'] = cdf['EmployeeName'].apply(lambda x: x.split()[0] + " " +x.split()[-1][:1] + ".")
```

```python
cdf.head(50)
```

```python
#We ended up with two problems. Data contains salaries of employees which is confidential information.
#Very likely surnames can be recognized by JobTitle and partially anonymized EmployeeName.
#Sa what we will now do is simply shuffling emplyees names. That should take care of both of our problems.
```

```python
from sklearn.utils import shuffle
import numpy as np
```

```python
np.random.shuffle(cdf['EmployeeName'].values)
```

```python
cdf.head(50)
```
